```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

#### **Node-Centric Graph**

```python
# Data for each graph construction method and model
models = ['GCN', 'GAT', 'GraphSAGE', 'Graphomer']
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-score', 'AUC-ROC']

node_centric_graph = {
    'GCN': [0.6435, 0.7013, 0.5038, 0.5864, 0.6519 ],
    'GAT': [0.6026, 0.6882, 0.3797, 0.4894, 0.6000],
    'GraphSAGE': [0.7788, 0.7691, 0.7988, 0.7837, 0.8550],
    'Graphomer': [0.8240, 0.8264, 0.8216, 0.8240, 0.9014]
}

# Create a bar chart for the results
x = np.arange(len(metrics))  # Metric indices
width = 0.2  # Bar width

fig, ax = plt.subplots(figsize=(12, 6))

# Plotting each model's metrics with values on top of bars
for i, (model, results) in enumerate(node_centric_graph.items()):
    bars = ax.bar(x + i * width, results, width, label=model)
    for bar, value in zip(bars, results):
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                f'{value:.3f}', ha='center', va='bottom', fontsize=10)

# Adding details to the chart
ax.set_xlabel('Metrics', fontsize=12)
ax.set_ylabel('Values', fontsize=12)
ax.set_title('Performance Metrics for Node-Centric Graph Models', fontsize=12)
ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(metrics, fontsize=10)
ax.legend(title="Models", fontsize=10, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=4)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### Heat Map comparing all Model Results

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Data for each graph construction method and model
models = ['GCN', 'GAT', 'GraphSAGE', 'Graphomer']
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-score', 'AUC-ROC']

node_centric_graph = {
    'GCN': [0.6435, 0.7013, 0.5038, 0.5864, 0.6519],
    'GAT': [0.6026, 0.6882, 0.3797, 0.4894, 0.6000],
    'GraphSAGE': [0.7788, 0.7691, 0.7988, 0.7837, 0.8550],
    'Graphomer': [0.8240, 0.8264, 0.8216, 0.8240, 0.9014],
}

edge_centric_graph = {
    'GCN': [0.9345, 0.9326, 0.9757, 0.9537, 0.9550],
    'GAT': [0.9572, 0.9843, 0.9533, 0.9685, 0.9830],
    'GraphSAGE': [0.9994, 0.9994, 0.9998, 0.9996, 1.0000],
    'Graphomer': [0.9994, 0.9994, 0.9997, 0.9996, 1.0000],
}

stratified_heterogeneous_graph = {
    'GCN': [0.6451, 0.6741, 0.5607, 0.6122, 0.6515],
    'GAT': [0.6448, 0.6837, 0.5376, 0.6019, 0.5969],
    'GraphSAGE': [0.9991, 0.9986, 0.9997, 0.9991, 0.9999],
    'Graphomer': [0.9994, 0.9996, 0.9992, 0.9994, 1.0000],
}

temporal_graph = {
    'GCN': [0.5465, 0.5347, 0.7090, 0.6097, 0.5680],
    'GAT': [0.5408, 0.5425, 0.5157, 0.5288, 0.5506],
    'GraphSAGE': [0.8070, 0.8024, 0.8140, 0.8082, 0.8831],
    'Graphomer': [0.7983, 0.7870, 0.8175, 0.8020, 0.8786],
}

graphs = {
    'Node-Centric Graph': node_centric_graph,
    'Edge-Centric Graph': edge_centric_graph,
    'Stratified Heterogeneous Graph': stratified_heterogeneous_graph,
    'Temporal Graph': temporal_graph,
}

# Visualization
for graph_name, graph_data in graphs.items():
    heatmap_data = pd.DataFrame(graph_data, index=metrics, columns=models)
    heatmap_data = heatmap_data.astype(float)  # Convert data to float

    plt.figure(figsize=(10, 6))
    sns.heatmap(
        heatmap_data, annot=True, fmt=".4f", cmap="YlGnBu", cbar=True,
        linewidths=0.5, linecolor='white'
    )
    plt.title(f"Model Performance on {graph_name}", fontsize=14, weight='bold')
    plt.ylabel('Metrics', fontsize=12)
    plt.xlabel('Models', fontsize=12)
    plt.tight_layout()
    plt.show()
```
